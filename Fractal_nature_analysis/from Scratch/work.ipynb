```python
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Function to tokenize lyrics into words
def tokenize_text(text):
    tokens = word_tokenize(text)  # Use NLTK's word_tokenize function
    return tokens

# Apply tokenization to 'Cleaned Line' column
df_1['Tokens'] = df_1['Line'].apply(tokenize_text)

# Display tokenized data
df_1[['Line', 'Tokens']]

```

```python
# Apply tokenization to 'Cleaned Line' column
df_2['Tokens'] = df_2['Line'].apply(tokenize_text)

# Display tokenized data
df_2[['Line', 'Tokens']]
```

```python
# Download NLTK resources if not already downloaded
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize stopwords and stemmer/lemmatizer
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Function to remove stop words, perform stemming, or lemmatization
def preprocess_text(tokens):
    # Remove stop words
    tokens = [word for word in tokens if word not in stop_words]
    
    # Perform stemming or lemmatization (choose one)
    # Example with stemming
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    
    # Example with lemmatization
    # lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    return stemmed_tokens  # Return stemmed or lemmatized tokens
```

```python
# Apply preprocessing to 'Tokens' column
df_1['Processed Tokens'] = df_1['Tokens'].apply(preprocess_text)

# Display preprocessed data
df_1[['Tokens', 'Processed Tokens']]
```

```python
# Apply preprocessing to 'Tokens' column
df_2['Processed Tokens'] = df_2['Tokens'].apply(preprocess_text)

# Display preprocessed data
df_2[['Tokens', 'Processed Tokens']]
```

```python
from transformers import BertTokenizer, BertModel

# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Tokenize and encode lyrics
tokenized_lyrics = df_1['Line'].apply(lambda x: tokenizer.encode(x, add_special_tokens=True))
```

```python
import torch

# Pad tokenized sequences to the same length
max_len = max(len(tokens) for tokens in tokenized_lyrics)
padded_lyrics = torch.tensor([tokens + [0]*(max_len - len(tokens)) for tokens in tokenized_lyrics])

# Define attention masks to differentiate real tokens from padding
attention_masks = torch.where(padded_lyrics != 0, torch.tensor(1), torch.tensor(0))

# Ensure tensors are in GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
padded_lyrics = padded_lyrics.to(device)
attention_masks = attention_masks.to(device)
```

```python
# Set the model to evaluation mode
model.eval()

# Move model to appropriate device
model = model.to(device)

# Forward pass through the model
with torch.no_grad():
    outputs = model(input_ids=padded_lyrics, attention_mask=attention_masks)

# Extract the hidden states or pooled output
hidden_states = outputs.last_hidden_state  # or outputs.pooler_output for BERT
```

```python
# Assuming 'outputs' from previous inference step
hidden_states = outputs.last_hidden_state  # Extract hidden states

# You can average the hidden states to get a pooled representation of the lyrics
pooled_output = torch.mean(hidden_states, dim=1)  # Example of pooling over tokens

# Convert tensors to numpy arrays for further analysis if needed
pooled_output = pooled_output.cpu().numpy()
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Perform dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
embeddings_tsne = tsne.fit_transform(pooled_output)

# Plot the embeddings
plt.figure(figsize=(10, 8))
plt.scatter(embeddings_tsne[:, 0], embeddings_tsne[:, 1], alpha=0.7)
plt.title('t-SNE Visualization of Lyrical Embeddings')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Example: Calculate cosine similarity between first and second lyric embeddings
similarity_score = cosine_similarity([pooled_output[0]], [pooled_output[1]])
print(f"Cosine Similarity Score: {similarity_score[0][0]}")
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize

# Example function to tokenize and tag documents
def tokenize_and_tag(data):
    tagged_data = [TaggedDocument(words=word_tokenize(row['Line'].lower()), tags=[row['Subdivision Title']]) for idx, row in data.iterrows()]
    return tagged_data

# Tokenize and tag documents
tagged_data = tokenize_and_tag(df_1)

# Train Doc2Vec model
model = Doc2Vec(tagged_data, vector_size=100, window=2, min_count=1, epochs=100)

# Example: Infer vector for a specific subdivision (e.g., 'Chorus')
vector_chorus = model.infer_vector(['staza 3'])

# Example: Find similar subdivisions (verses) based on vector similarity
similar_subdivisions = model.docvecs.most_similar('stanza 3')
print(similar_subdivisions)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example function to create recurrence plot
def recurrence_plot(data):
    # Convert lyrical data to binary matrix
    binary_matrix = np.zeros((len(data), len(data)))
    for i in range(len(data)):
        for j in range(len(data)):
            if data[i] == data[j]:
                binary_matrix[i, j] = 1
    
    # Plot recurrence plot
    plt.imshow(binary_matrix, cmap='binary', origin='lower')
    plt.title('Recurrence Plot of Lyrical Patterns')
    plt.xlabel('Time')
    plt.ylabel('Time')
    plt.show()

# Example: Create recurrence plot for 'Line' column in df
recurrence_plot(df_1['Line'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation as LDA

# Example function to apply LDA
def apply_lda(data):
    # Initialize CountVectorizer
    vectorizer = CountVectorizer(lowercase=True, stop_words='english')
    X = vectorizer.fit_transform(data['Line'])
    
    # Apply LDA
    lda = LDA(n_components=5, random_state=42)
    lda.fit(X)
    
    # Display topics
    for topic_idx, topic in enumerate(lda.components_):
        print(f"Topic {topic_idx + 1}:")
        print([vectorizer.get_feature_names()[i] for i in topic.argsort()[-10:]])
        print()

# Apply LDA on 'Line' column
apply_lda(df_1)
```
